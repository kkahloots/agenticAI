# Level 3 — Functional Agent (Agentic Architecture)

This notebook demonstrates the **Agentic AI** implementation of Level 3 Functional capabilities.

## Key Features

- **Planning**: Multi-step reasoning and task decomposition
- **Dynamic Tool Discovery**: Registry-based tool selection
- **Dynamic SQL Generation**: Safe query generation from natural language
- **MCP Tools**: Model Context Protocol abstraction layer
- **Memory Layer**: Tracks plans, tools, compliance decisions
- **Evaluation**: Result validation with replanning capability

## Two Versions Per Use Case

Each use case has:
- **Version A (Deterministic)**: Fixed tool paths, baseline behavior
- **Version B (Dynamic)**: Dynamic planning, tool discovery, SQL generation

## Use Cases

1. **UC1**: Lead Scoring
2. **UC2**: Customer Enrichment
3. **UC3**: Next-Best-Action (NBA) Recommendation
4. **UC4**: Consent-Gated Notification
5. **UC5**: Identity Gate
6. **UC6**: Bulk Campaign Targeting
7. **UC7**: Return Risk Intervention
8. **UC8**: Campaign Results Dashboard
9. **UC9**: Guardrails

## UC1: Lead Scoring — Identify Top Prospects

### Version A: Deterministic Agentic Baseline
Fixed tool path: `score_leads` tool

```python
display_title("UC1-A: Lead Scoring (Deterministic)", emoji="🎯")

# Execute with deterministic mode
result_det = api.execute_uc1_lead_scoring(
    offer_code="PROMO-PREMIUM-MEMBERSHIP",
    top_n=10,
    mode="deterministic"
)

# Show agentic metadata
show_agentic_result(result_det)
show_agent_plan(result_det)
show_agent_path(result_det)

# Show business result (output compatibility)
final = result_det['final_result']['result']
display_lead_scoring_results(final)
```

### Comparison: Nonagent vs Deterministic vs Dynamic

## UC7: Return Risk Intervention

```python
display_title("UC7-A: Return Risk (Deterministic)", emoji="⚠️")

result_det = api.execute_uc7_return_risk(threshold=0.7, mode="deterministic")
show_agentic_result(result_det)
show_agent_plan(result_det)

# Show business result
customers = json.loads(Path("../../data/customers.json").read_text())
at_risk = [c for c in customers if c.get("return_risk", 0) > 0.7 
           and c.get("identity_status") in ("verified", "pending")
           and c.get("consent_flags", {}).get("marketing", False)]
at_risk.sort(key=lambda x: x["return_risk"], reverse=True)
display_return_risk_intervention(at_risk, send_notification)
```
