## What is Level 1?

The **Level 1 Knowledge & Retrieval System** is the system's information retrieval layer. It answers questions by searching across customer databases, email archives, CRM notes, and policy documents — all in plain English, no SQL required.

Think of it as an intelligent search engine that understands meaning, not just keywords. Instead of returning raw documents, it synthesizes clear, cited answers.

### Key Capabilities

- **Query Understanding**: Restates user questions and extracts keywords/entities for better search precision
- **Semantic Search**: Finds relevant documents using vector embeddings (Chroma/Pinecone), understanding meaning not just keywords
- **Multi-Query Retrieval**: Generates multiple reformulations to improve recall from different angles
- **Answer Synthesis**: Uses LLM to create clear, cited answers from retrieved documents (RAG)
- **Memory Integration**: Stores requests and responses for conversation context in follow-up questions
- **PII Masking**: Automatically redacts sensitive information (emails, phone numbers, names) before returning results

---

## How it works (in plain terms)

```
Your question
     │
     ▼
Query Understanding: Extract keywords and entities
     │
     ▼
Multi-Query Retrieval: Generate multiple reformulations
     │
     ▼
Semantic Search: Find top documents from vector store
     │
     ▼
Context Assembly: Select and organize retrieved chunks
     │
     ▼
Answer Generation: LLM synthesizes answer with citations
     │
     ▼
PII Masking: Redact sensitive information
     │
     ▼
Memory Update: Store request and response for context
     │
     ▼
Answer returned with sources
```
